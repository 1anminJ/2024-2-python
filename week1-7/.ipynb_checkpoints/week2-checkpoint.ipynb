```python
import pandas as pd
# dic_exam = {
#     key:value,
#     key2:value2,
#     key3:value3
# }
dic_exam = {
    "name" : ["홍", "김", "이"],
    "english" : [100, 200, 300],
    "math" : [70, 80, 90]
}
df_exam = pd.DataFrame(dic_exam)
df_exam
# type(df_exam)  # 실행 결과 : pandas.core.frame.DataFrame
```

```python
# series(1차원 배열)
type(df_exam['english']) # pandas.core.series.Series
df_exam['english']
```

```python
df_exam[['english']] # 대괄호를 두 개 작성하면 데이터 프레임으로 볼 수 있음(미관상 좋음)
```

```python
print("sum : ", sum(df_exam['english']))
print("min : ", min(df_exam['english']))
print("max : ", max(df_exam['english']))
```

```python
# 평균 - built in function 사용
print("평균 : ", sum(df_exam['english'])/len(df_exam['english']))
# len은 열의 개수를 세어주기에 갯수를 나타냄
```

```python
# 혼자서 해보기
dic_fruits = {
    "제품" : ["사과", "딸기", "수박"],
    "가격" : [1800, 1500, 3000],
    "판매량" : [24, 38, 13]
}
df_fruits = pd.DataFrame(dic_fruits)
df_fruits
```

```python
# 과일 가격 평균
print("과일 가격 평균 : ",sum(df_fruits['가격'])/len(df_fruits['가격']))
# 과일 판매량 평균
print("과일 판매량 평균 : ", sum(df_fruits['판매량'])/len(df_fruits['판매량'])
```
