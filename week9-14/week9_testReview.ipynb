```python
# 실기 시험 풀이
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family' : 'AppleGothic'})
d = pd.read_csv('week9_movieData.csv')
```

```python
# d.head()
# d.shape # 32244개의 데이터가 있음
# d.info()
# d.describe()
d.head(1)
```

```python
# 문제1. 가장 높은 평균평점을 받은 영화 top 10을 골라라
d.groupby('영화제목').agg(평균평점 = ('평점', 'mean'))\
    .sort_values('평균평점', ascending=False).head(10)
```

```python
# 문제2. 가장 많은 사람이 평가한 top 5를 골라라
d.groupby('영화제목').agg(평가한사람수 = ('영화제목', 'count')).sort_values('평가한사람수', ascending=False).head()
```

```python
# 문제3. 600명 이상 평가를 한 영화 중에 평균 평점이 높은 순서대로 나열
d.groupby('영화제목').agg(평가한사람수 = ('영화제목', 'count'), 평균평점 = ('평점', 'mean'))\
    .query('평가한사람수 >= 600').sort_values('평균평점', ascending=False)
```

```python
# 문제4. 영화 "Toy Story (1995)" 를 평가한 사람들의 연령의 분포를 그래프로 나타내라
result = d.query('영화제목 == "Toy Story (1995)"')
# sns.boxplot(data = result, y = '연령')
# sns.histplot(data = result, x = '연령')
sns.countplot(data = result, x = '연령')
```

```python
# 문제5. 여자들이 제일 높은 평균 평점을 준 영화 제목 10개를 골라라
d.query('성별 == "F"').groupby('영화제목').agg(평균평점 = ('평점', 'mean')).sort_values('평균평점', ascending=False).head(10)
```
